```python
import pandas as pd 
data = pd.read_csv('solved_ac_data.csv')
print(data['닉네임'].nunique())
unique_users = data['닉네임'].unique()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# ChromeDriver 설정
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  # 브라우저를 최대화
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

wait = WebDriverWait(driver, 10)

# 유저 검색 및 필터링 함수
def check_user_status(user_id):
    url = "https://www.acmicpc.net/status"
    driver.get(url)

    try:
        # 검색창이 나타날 때까지 기다리기
        search_box = wait.until(EC.presence_of_element_located((By.NAME, "user_id")))
        
        # 검색창에 유저 ID 입력
        search_box.clear()
        search_box.send_keys(user_id)
        search_box.submit()
        
        # time.sleep(1)  # 페이지 로드 대기
        
        # 언어 필터링 (C99, Python 3, C++17, PyPy3, Java 11만 허용)
        try:
            first_language = driver.find_element(By.XPATH, "//tr[contains(@id, 'solution-')]/td[7]")  # 7번째 열이 언어
            allowed_languages = ['Python 3', 'C99', 'C++17', 'PyPy3', 'Java 11', 'Java 9', 'Java 8', 'Java 15', 'C++17 (Clang)', 'C++20', 'C++14', 'C99 (Clang)']
            if first_language.text not in allowed_languages:
                print(f"{user_id}: 부계정 의심 (허용되지 않는 언어 {first_language.text})")
                return "sub_account"
        except Exception as e:
            print(f"{user_id}: 언어 정보를 찾을 수 없음. {e}")
        
        # 제출 시간 필터링 (제출 시간이 1년 이상일 경우)
        try:
            submission_time = driver.find_element(By.XPATH, "//tr[contains(@id, 'solution-')]/td[9]")  # 9번째 열이 제출 시간
            if '1년' in submission_time.text:
                print(f"{user_id}: 1년 이상 제출하지 않은 유저")
                return "inactive"
        except Exception as e:
            print(f"{user_id}: 제출 시간 정보를 찾을 수 없음. {e}")
    
    except Exception as e:
        print(f"유저 {user_id} 검색 중 오류 발생: {e}")
    
    return "active"

# 유니크 유저 필터링 실행
filtered_users = []
try:
    for idx, user in enumerate(unique_users):  # unique_users 리스트는 외부에서 정의된다고 가정
        result = check_user_status(user)
        if result != "active":
            filtered_users.append(user)
            print(f"필터링된 유저: {user}")

except Exception as e:
    print(f"에러 발생: {e}")

# Selenium 드라이버 종료
driver.quit()

# 최종 필터링된 유저 출력
print(f"필터링된 유저 리스트: {filtered_users}")
```

```python
# 필터링된 유저 데이터를 데이터에서 삭제
filtered_data = data[~data['닉네임'].isin(filtered_users)]

# 새로운 CSV 파일로 저장
filtered_data.to_csv('filtered_solved_ac_data(Gold II ~ Platinum I).csv', index=False, encoding='utf-8-sig')

print("필터링된 데이터를 저장했습니다.")
```
